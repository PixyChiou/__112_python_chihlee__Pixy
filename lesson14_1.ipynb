```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scroes_df = pd.DataFrame(scroes,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
scroes_df
sum_value = scroes_df.sum(axis=1) #當axis=0時，代表每一縱行去做運算 當axis=1時 代表每一橫列去做
mean_value = scroes_df.mean(axis=1) #mean平均數

scroes_df['總分'] = sum_value
scroes_df['平均'] = mean_value
scroes_df
```

```python
scroes_df.sort_values(by='總分',ascending=False).iloc[:5]# ascending排序,預設為true,由小到大,要由大到小故要指定為false , iloc~前5名
scroes_df.sort_values(by='總分',ascending=False).iloc[-5:]#後5名
```

```python
#平均小於70 使用subscritp[],過濾資料
scroes_df[scroes_df['平均'] < 70]
```

```python
#使用subscritp[]的搜尋方式,過濾資料
#平均小於80,且歷史不及格
scroes_df[(scroes_df['平均'] < 80) & (scroes_df['歷史'] < 60 )]
```

```python
#使用query()實體方法的搜尋
scroes_df.query('平均<80')
```

```python
#使用query()實體方法的搜尋
scroes_df.query('平均<80 and 歷史<60')
```
