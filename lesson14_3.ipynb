```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scroes_df = pd.DataFrame(scroes,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
scroes_df
sum_value = scroes_df.sum(axis=1) #當axis=0時，代表每一縱行去做運算 當axis=1時 代表每一橫列去做
mean_value = scroes_df.mean(axis=1) #mean平均數

scroes_df['總分'] = sum_value
scroes_df['平均'] = mean_value
```

```python
scroes_df
```

```python
# precision=2 指小數點二位數
# highlight_between 指50-59分反黃
# 用\來表示 這個和下一行是同一行
scroes_df.style\
    .format(precision=2)\
    .highlight_between(left=50,right=59)\
    .to_excel('abc.xlsx')
```

```python
sc2 = scroes_df.sort_values(by='總分',ascending=False)
sc2.style\
    .format(precision=2)\
    .highlight_between(left=50,right=59)\
    .to_excel('abc2.xlsx')
```
